```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
headers = {"User-Agent": 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.81 Safari/537.36'}
```

```python
webpage = requests.get('https://www.bproperty.com/rent/dhaka/residential/apartments/', headers=headers).text
```

```python
soup = BeautifulSoup(webpage, 'lxml')
```

```python
soup.prettify()
```

```python
house = soup.find_all('div', class_ = 'ListingCell-AllInfo ListingUnit')
```

```python
len(house)
```

```python
heading = []
location = []
rent_per_month = []
bedRooom = []
bathRooom = []
sqft = []
for i in house:
    heading.append(i.find('h3').text.strip())
    location.append(i.find('span', class_ ="ListingCell-KeyInfo-address-text").text.strip())
    rent_per_month.append(i.find('span', class_ ="PriceSection-FirstPrice").text.strip())
    try:
        bedRooom.append(i.find_all('span', class_ ="KeyInformation-value_v2 KeyInformation-amenities-icon_v2")[0].text.strip())
    except:
        bedRooom.append(np.nan)
    try:
        bathRooom.append(i.find_all('span', class_ ="KeyInformation-value_v2 KeyInformation-amenities-icon_v2")[1].text.strip())
    except:
        bathRooom.append(np.nan)

    try:
        sqft.append(i.find_all('div', class_ ="KeyInformation-description_v2")[2].text.strip())
    except:
        sqft.append(np.nan)

d = {'heading': heading, 'location': location, 'rent per month': rent_per_month, 'Bedrooom': bedRooom, 'Bathrooom': bathRooom, 'Area-sqft': sqft}
df = pd.DataFrame(d)
```

```python
final =pd.DataFrame()
for j in range(1,51):
  url ='https://www.bproperty.com/rent/dhaka/residential/apartments/?page={}'.format(j)
  headers = {"User-Agent": 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.81 Safari/537.36'}
  webpage = requests.get(url, headers=headers).text
  soup = BeautifulSoup(webpage, 'lxml')
  house = soup.find_all('div', class_ = 'ListingCell-AllInfo ListingUnit')
  heading = []
  location = []
  rent_per_month = []
  bedRooom = []
  bathRooom = []
  sqft = []
  for i in house:
      heading.append(i.find('h3').text.strip())
      location.append(i.find('span', class_ ="ListingCell-KeyInfo-address-text").text.strip())
      try:
          rent_per_month.append(i.find('span', class_ ="PriceSection-FirstPrice").text.strip())
      except:
          rent_per_month.append(np.nan)
      try:
          bedRooom.append(i.find_all('span', class_ ="KeyInformation-value_v2 KeyInformation-amenities-icon_v2")[0].text.strip())
      except:
          bedRooom.append(np.nan)
      try:
          bathRooom.append(i.find_all('span', class_ ="KeyInformation-value_v2 KeyInformation-amenities-icon_v2")[1].text.strip())
      except:
          bathRooom.append(np.nan)

      try:
          sqft.append(i.find_all('div', class_ ="KeyInformation-description_v2")[2].text.strip())
      except:
          sqft.append(np.nan)

  d = {'heading': heading, 'location': location, 'rent per month': rent_per_month, 'Bedrooom': bedRooom, 'Bathrooom': bathRooom, 'Area-sqft': sqft}
  df = pd.DataFrame(d)

  final = pd.concat([final, df], ignore_index=True)
```
